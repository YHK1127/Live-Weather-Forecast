```python
from tkinter import *
import requests
import json
from datetime import datetime
from PIL import Image, ImageTk
from io import BytesIO

#Initialize Window
root = Tk()
root.geometry("400x400") #size of the window by default
root.resizable(0,0) #to make the window size fixed
root.title("Weather App")
# Set the background color using a named color or hexadecimal code
root.configure(bg="lightblue")
 
#Functions to fetch and display weather info
city_value = StringVar()
 
def time_format_for_location(utc_with_tz):
    local_time = datetime.utcfromtimestamp(utc_with_tz)
    return local_time.time()
 
city_value = StringVar()
 
def get_weather_icon_url(icon_code):
    return f"http://openweathermap.org/img/wn/{icon_code}.png"
 
def showWeather():
    #Enter your API key copied from the OpenWeatherMap dashboard
    api_key = "xyz2334455yyhhnfcdd"  #sample API
 
    # Get city name from user input field
    city_name=city_value.get()
 
    # API url
    weather_url = 'http://api.openweathermap.org/data/2.5/weather?q=' + city_name + '&appid='+api_key
 
    # Get the response from fetched url
    response = requests.get(weather_url)
 
    # changing response from json to python readable 
    weather_info = response.json()
 
 
    tfield.delete("1.0", "end")   #to clear the text field for every new output
 
    if weather_info['cod'] == 200:
        kelvin = 273 # value of kelvin

        #Storing the fetched values of weather of a city
 
        temp = int(weather_info['main']['temp'] - kelvin)                                     #converting default kelvin value to Celcius
        feels_like_temp = int(weather_info['main']['feels_like'] - kelvin)
        pressure = weather_info['main']['pressure']
        humidity = weather_info['main']['humidity']
        wind_speed = weather_info['wind']['speed'] * 3.6
        sunrise = weather_info['sys']['sunrise']
        sunset = weather_info['sys']['sunset']
        timezone = weather_info['timezone']
        cloudy = weather_info['clouds']['all']
        description = weather_info['weather'][0]['description']
 
        sunrise_time = time_format_for_location(sunrise + timezone)
        sunset_time = time_format_for_location(sunset + timezone)
#assigning Values to our weather variable, to display as output
         
        weather = f"\nWeather of: {city_name}\nTemperature (Celsius): {temp}°\nFeels like in (Celsius): {feels_like_temp}°\nPressure: {pressure} hPa\nHumidity: {humidity}%\nSunrise at {sunrise_time} \nSunset at {sunset_time}\nCloud: {cloudy}%\nInfo: {description}"
 
        # Fetch the weather icon from URL
        icon_code = weather_info['weather'][0]['icon']
        weather_icon_url = get_weather_icon_url(icon_code)
        response_icon = requests.get(weather_icon_url)
 
        # Check if the request was successful
        if response_icon.status_code == 200:
            icon_data = response_icon.content
 
            # Convert the icon data to an image
            icon_image = Image.open(BytesIO(icon_data))
            icon_image = icon_image.resize((75, 75))  # Resize the icon to fit the window
            icon_photo = ImageTk.PhotoImage(icon_image)
 
            # Clear previous weather icon
            weather_icon_label.config(image=None)
 
            # Update the weather icon label with the new icon
            weather_icon_label.config(image=icon_photo)
            weather_icon_label.image = icon_photo
 
    else:
        weather = f"\n\tWeather for '{city_name}' not found!\n\tKindly Enter valid City Name !!"
 
        # Clear previous weather icon
        weather_icon_label.config(image=None)
 
    tfield.insert(INSERT, weather)   #to insert or send value in our Text Field to display output
 
 
 #Frontend part of code - Interface
 
city_head= Label(root, text = 'Enter City Name', font = 'TimesNewRoman 12 bold',bg='purple',fg='white').pack(pady=10) #to generate label heading
 
inp_city = Entry(root, textvariable = city_value,  width = 24, font='TimesNewRoman 14 bold').pack()
 
Button(root, command = showWeather, text = "Check Weather", font="Arial 10", bg='purple', fg='white', activebackground="teal", padx=5, pady=5 ).pack(pady= 20)
 
#to show output
 
weather_now = Label(root, text = "The Weather is:", font = 'arial 12 bold',bg='purple',fg='white').pack(pady=10)
 
tfield = Text(root, width=44, height=10)
tfield.pack()
 
# Create a label to display the weather icon
weather_icon_label = Label(root)
weather_icon_label.pack()
 
root.mainloop()
```
